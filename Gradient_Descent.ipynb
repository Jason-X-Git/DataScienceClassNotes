# Example 1
## $f(x) = x^2 + x +1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# Plot
plt.xlim(-3, 3)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
```

# Slope & Derivationves

```python
def df(x):
    return (2*x + 1)
```

```python
# Plot function and derivation side by side
plt.figure(figsize=[15,5])

# 1 Chart: Cost Function
plt.subplot(1, 2,1)
plt.title('Cost Function', fontsize=16)
plt.xlim(-3, 3)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()
plt.plot(x_1, f(x_1), color='red', linewidth=3)

# 2 chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of cost function', fontsize=16)
plt.xlim(-3, 3)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()

plt.plot(x_1, df(x_1), color='blue', linewidth=3)
```

## Python Loops & Gradient Descent

```python
# Python For Loop
for n in range(5):
    print('Hello World', n)
print('End of loop')
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0000000000001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(100000):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print('Ran {} steps'.format(n))
        break
    
print('Local minium occurs at: ', new_x)
print('Slope value is: ', df(new_x))
print('Cost value is: ', f(new_x))
```

```python
# Superimposes the gradient descent calculations on plots
plt.figure(figsize=[15,5])

# 1 Chart: Cost Function
plt.subplot(1, 2,1)
plt.title('Cost Function', fontsize=16)
plt.xlim(-3, 3)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()
plt.plot(x_1, f(x_1), color='red', linewidth=5, alpha=0.5)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='blue', s=100, alpha=0.6)

# 2 chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of cost function', fontsize=16)
plt.xlim(-3, 3)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()

plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.5)
```

# Example 4 - Data Viz with 3D Charts
## Minimise $$f(x,y) = \frac{1}{3^{(-x^2- y^2)} + 1}$$

```python
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1/ (r+1)
```

```python
# Make our x and y data
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)
```

```python
type(x_4), x_4.shape
```

```python
# make 1d array to 2d array
x_4, y_4 = np.meshgrid(x_4, y_4)
```

```python
x_4.shape, y_4.shape
```

```python
# Grenerating 3d Plot
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.set_xlabel('X', fontsize=20)
ax.set_ylabel('Y', fontsize=20)
ax.set_zlabel('f(x, y) Cost', fontsize=20)

ax.plot_surface(x_4, y_4,  f(x_4, y_4), alpha=0.8, cmap=cm.plasma)
```
